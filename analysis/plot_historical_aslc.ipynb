```python
import numpy as np
import xarray as xr
import cmocean
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from analysis_io import open_aslc_range, open_aslc_amax
from analysis import wrap_argmonth_diff, compute_circ_mean
from mask_regions import mask_inland_seas, mask_poles
from make_cartopy_maps import add_global_map_subplot, add_global_map_of_months_subplot, get_discrete_cmap
```

Open CMIP6 ASLC data:

```python
#open historical ASLC range and amax:
aslc_range = open_aslc_range('gs://leap-persistent/timh37/CMIP6/aslc_1x1',{'lat':500,'lon':500,'period':1,'member_id':50,'experiment_id':4}).sel(period='hist')#.load()
aslc_range_tgs = open_aslc_range('gs://leap-persistent/timh37/CMIP6/aslc_psmsl',{'tg':1000,'period':2,'member_id':200}).sel(period='hist').load()

aslc_amax = open_aslc_amax('gs://leap-persistent/timh37/CMIP6/aslc_1x1',{'lat':500,'lon':500,'period':1,'member_id':50,'experiment_id':4}).sel(period='hist')#.load()
aslc_amax_tgs = open_aslc_amax('gs://leap-persistent/timh37/CMIP6/aslc_psmsl',{'tg':1000,'period':2,'member_id':200}).sel(period='hist').load()

#apply some filters:
#1): select experiment with most available members, on a model by model basis (if equal numbers, pick first)
availability = np.isfinite(aslc_range_tgs['zos_ibe']).any(dim=['id']).compute()
num_members_available = availability.sum(dim='member_id')
available_for_all_ssps = availability.all(dim='experiment_id').sum(dim='member_id')
i_ssp_with_most_mem = num_members_available.argmax(dim='experiment_id')

aslc_range = aslc_range.isel(experiment_id=i_ssp_with_most_mem,drop=True)
aslc_range_tgs = aslc_range_tgs.isel(experiment_id=i_ssp_with_most_mem,drop=True)

aslc_amax = aslc_amax.isel(experiment_id=i_ssp_with_most_mem,drop=True)
aslc_amax_tgs = aslc_amax_tgs.isel(experiment_id=i_ssp_with_most_mem,drop=True)

#2): mask out where less than 10 models are ocean
more_than_10_models = np.isfinite(aslc_range.zos_ibe).any(dim='member_id').sum(dim='source_id')>=10 #mask for gridded fields where less than 10 models have data

#3): mask out poles and inland seas
aslc_range = mask_poles( mask_inland_seas( aslc_range.where(more_than_10_models) ) ).load()
aslc_range_tgs = mask_poles( mask_inland_seas( aslc_range_tgs ) )

aslc_amax = mask_poles( mask_inland_seas( aslc_amax.where(more_than_10_models) ) ).load()
aslc_amax_tgs = mask_poles( mask_inland_seas( aslc_amax_tgs ) )
```

Open observation month means & compute range/amax:

```python
###LOAD DATA
#tide-gauge data
psmsl = xr.open_dataset('../month_means_1993_2022_psmsl.nc') #month means
psmsl = mask_poles(mask_inland_seas(psmsl))
psmsl = psmsl.where(np.isfinite(psmsl.isel(month=0,drop=True)),drop=True)

#satellite altimetry and era5-based ibe
sla = xr.open_dataset('../allsat_l4_duacs_sla_1993-2023_1x1.nc') #time series
#sla = xr.open_dataset('../allsat_l4_duacs_sla_1993-2023_1x1_coarsened.nc') #time series
ibe = xr.open_dataset('../monthly_ibe_1993-2023_1x1.nc') #time series

sla = mask_poles(mask_inland_seas(sla))
ibe = mask_poles(mask_inland_seas(ibe))

sla_ibe = sla['sla']+ ibe['ibe'].values

sla_month_means = sla.sla.groupby(sla.time.dt.month).mean()
ibe_month_means = ibe.ibe.groupby(ibe.time.dt.month).mean()
sla_ibe_month_means = sla_ibe.groupby(sla_ibe.time.dt.month).mean()

#satellite altimetry and era5-based ibe at tide-gauges
sla_tgs = xr.open_dataset('../duacs-0.125deg_P1M-m_202411_at_tgs.nc').sel({'id':psmsl.id})
ibe_tgs = xr.open_dataset('../monthly_ibe_1993-2023_at_psmsl.nc').sel({'id':psmsl.id})

sla_ibe_tgs = sla_tgs['sla']+ ibe_tgs['ibe'].values
sla_tgs = sla_tgs.sel({'id':psmsl.id})

sla_month_means_tgs = sla_tgs.sla.groupby(sla_tgs.time.dt.month).mean()
ibe_month_means_tgs = ibe_tgs.ibe.groupby(ibe_tgs.time.dt.month).mean()
sla_ibe_month_means_tgs = sla_ibe_tgs.groupby(sla_ibe_tgs.time.dt.month).mean()


###COMPUTE RANGE AND AMAX
#tide gauges
psmsl_range = 0.1*(psmsl.monmean.max(dim='month') - psmsl.monmean.min(dim='month'))#[cm]
psmsl_amax = 1+psmsl.monmean.argmax(dim='month',skipna=False).where(np.isfinite(psmsl.monmean.isel(month=0,drop=True)))

psmsl_minus_ibe_range = 0.1*((psmsl.monmean-1000*ibe_month_means_tgs).max(dim='month') - (psmsl.monmean-1000*ibe_month_means_tgs).min(dim='month'))#[cm]
psmsl_minus_ibe_amax = 1+(psmsl.monmean-1000*ibe_month_means_tgs).argmax(dim='month',skipna=False).where(np.isfinite(psmsl.monmean.isel(month=0,drop=True)))

#satellite
sla_range = 100*(sla_month_means.max(dim='month') - sla_month_means.min(dim='month')) #[cm]
sla_range_tgs = 100*(sla_month_means_tgs.max(dim='month') - sla_month_means_tgs.min(dim='month')) #[cm]

ibe_range = 100*(ibe_month_means.max(dim='month') - ibe_month_means.min(dim='month')) #[cm]
ibe_range_tgs = 100*(ibe_month_means_tgs.max(dim='month') - ibe_month_means_tgs.min(dim='month')) #[cm]

sla_ibe_range = 100*(sla_ibe_month_means.max(dim='month') - sla_ibe_month_means.min(dim='month')) #[cm]
sla_ibe_range_tgs = 100*(sla_ibe_month_means_tgs.max(dim='month') - sla_ibe_month_means_tgs.min(dim='month')) #[cm]

sla_amax = 1+sla_month_means.argmax(dim='month',skipna=False) + 0*sla_month_means.isel(month=0).values
sla_amax_tgs = 1+sla_month_means_tgs.argmax(dim='month',skipna=False) + 0*sla_month_means_tgs.isel(month=0).values

sla_ibe_amax = 1+sla_ibe_month_means.argmax(dim='month',skipna=False) + 0*sla_ibe_month_means.isel(month=0).values
sla_ibe_amax_tgs = 1+sla_ibe_month_means_tgs.argmax(dim='month',skipna=False) + 0*sla_ibe_month_means_tgs.isel(month=0).values

ibe_amax = 1+ibe_month_means.argmax(dim='month',skipna=False) + 0*ibe_month_means.isel(month=0).values
ibe_amax_tgs = 1+ibe_month_means_tgs.argmax(dim='month',skipna=False) + 0*ibe_month_means_tgs.isel(month=0).values
```

```python
#select cmip6 data only where observations are not masked out
aslc_range_tgs = aslc_range_tgs.sel({'id':psmsl.id}) 
aslc_amax_tgs = aslc_amax_tgs.sel({'id':psmsl.id})
```

Compute error statistics for ‘zos_ibe’:

```python
this_var='zos_ibe'

#compute error statistics:
#Per member:
range_mae = np.abs(aslc_range[this_var] - sla_ibe_range).mean(dim=('lon','lat'))
range_mape = 100*np.abs((aslc_range[this_var] - sla_ibe_range)/sla_ibe_range).mean(dim=('lon','lat'))

range_mae_tgs = np.abs(aslc_range_tgs[this_var] - psmsl_range).mean(dim=('id'))
range_mape_tgs = 100* np.abs((aslc_range_tgs[this_var] - psmsl_range)/psmsl_range).mean(dim=('id'))

amax_mae = np.abs(wrap_argmonth_diff(aslc_amax[this_var] - sla_ibe_amax)).mean(dim=('lon','lat'))
amax_mae_tgs = np.abs(wrap_argmonth_diff(aslc_amax_tgs[this_var] - psmsl_amax)).mean(dim=('id'))
                                                              
#For the ensemble mean:
range_mae_ensmean = np.abs(aslc_range[this_var].mean(dim='member_id').mean(dim='source_id') - sla_ibe_range).mean(dim=('lon','lat'))
range_mape_ensmean  = 100*np.abs((aslc_range[this_var].mean(dim='member_id').mean(dim='source_id') - sla_ibe_range)/sla_ibe_range).mean(dim=('lon','lat'))

range_mae_tgs_ensmean = np.abs(aslc_range_tgs[this_var].mean(dim='member_id').mean(dim='source_id') - psmsl_range).mean(dim=('id'))
range_mape_tgs_ensmean = 100* np.abs((aslc_range_tgs[this_var].mean(dim='member_id').mean(dim='source_id') - psmsl_range)/psmsl_range).mean(dim=('id'))

amax_mae_ensmean = np.abs(wrap_argmonth_diff(compute_circ_mean(compute_circ_mean(aslc_amax[this_var],dim='member_id'),dim='source_id') - sla_ibe_amax)).mean(dim=('lon','lat'))
amax_mae_tgs_ensmean = np.abs(wrap_argmonth_diff(compute_circ_mean(compute_circ_mean(aslc_amax_tgs[this_var],dim='member_id'),dim='source_id') - psmsl_amax)).mean(dim=('id'))

#####Altimetry v.s. TGs
sla_ibe_range_mae_tgs = np.abs(sla_ibe_range_tgs - psmsl_range).mean(dim=('id'))
sla_ibe_range_mape_tgs = 100*np.abs((sla_ibe_range_tgs - psmsl_range)/psmsl_range).mean(dim=('id'))

sla_ibe_amax_mae_tgs = np.abs(wrap_argmonth_diff(sla_ibe_amax_tgs - psmsl_amax)).mean(dim=('id'))
```
